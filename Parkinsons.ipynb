```python
# Feature Target Split
x = df.drop('status', axis=1)
y = df['status']
```

```python
# Scaling the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Model Training - Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(x_train, y_train)
```

```python
# Predicting on test set
lr_pred = lr.predict(x_test)
```

```python
# Accuracy Score
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, lr_pred)
print(f"Accuracy of Logistic Regression: {accuracy:.2f}")
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=300,       # number of trees
    random_state=42,
    class_weight="balanced") # handles imbalance (important for your dataset)
rf.fit(x_train, y_train)
```

```python
# Predictions
rf_pred = rf.predict(x_test)
rf_probs = rf.predict_proba(x_test)[:, 1]
# Accuracy Score for Random Forest
print(f"Accuracy  : {accuracy_score(y_test, rf_pred):.3f}")
```

```python
# Metrics for both models
from sklearn.metrics import classification_report, confusion_matrix
# Classification report for Logistic Regression
print("\nClassification Report for Logistic Regression:")
print(classification_report(y_test, lr_pred))
# Confusion Matrix for Logistic Regression
print("\nConfusion Matrix for Logistic Regression:")
print(confusion_matrix(y_test, lr_pred))
# Classification report for Random Forest
print("\nClassification Report for Random Forest:")
print(classification_report(y_test, rf_pred))
# Confusion Matrix for Random Forest
print("\nConfusion Matrix for Random Forest:")
print(confusion_matrix(y_test, rf_pred))

# Plotting confusion matrix for Random Forest
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay
cm = confusion_matrix(y_test, rf_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Random Forest")
plt.show()
# plottinf confusion matrix for Logistic Regression
cm_lr = confusion_matrix(y_test, lr_pred)
disp_lr = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=lr.classes_)
disp_lr.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Logistic Regression")
plt.show()
```

```python
# ROC Curve for Random Forest
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, rf_probs)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()

# ROC accuracy
print(f"ROC AUC for Random Forest: {roc_auc:.2f}")
```

```python
# 🔎 Random Forest Hyperparameter Tuning with GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    "n_estimators": [100, 200, 300, 500],   # number of trees
    "max_depth": [None, 5, 10, 20],         # limit tree depth
    "min_samples_split": [2, 5, 10],        # minimum samples to split
    "min_samples_leaf": [1, 2, 4],          # minimum samples at leaf
    "class_weight": [None, "balanced"]      # handle imbalance
}

# Initialize model
rf = RandomForestClassifier(random_state=42)

# Grid search with 5-fold cross-validation
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    scoring="accuracy",   # could also try "f1" or "roc_auc"
    cv=5,
    n_jobs=-1,            # use all CPU cores
    verbose=2
)

# Fit
grid_search.fit(x_train, y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)

# Use the best model
best_rf = grid_search.best_estimator_

# Evaluate on test set
rf_pred = best_rf.predict(x_test)
rf_probs = best_rf.predict_proba(x_test)[:, 1]

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

print("\n=== Test Set Evaluation ===")
print("Accuracy :", accuracy_score(y_test, rf_pred))
print("ROC-AUC  :", roc_auc_score(y_test, rf_probs))
print("\nClassification Report:\n", classification_report(y_test, rf_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_pred))

#plotting confusion matrix for best Random Forest model
cm_best_rf = confusion_matrix(y_test, rf_pred)
disp_best_rf = ConfusionMatrixDisplay(confusion_matrix=cm_best_rf, display_labels=best_rf.classes_)
disp_best_rf.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Best Random Forest Model")
plt.show()
```
