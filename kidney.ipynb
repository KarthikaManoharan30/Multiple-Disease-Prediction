```python
import pandas as pd
import numpy as np
```

```python
kidney_df = pd.read_csv(r'D:\Projects\Multi_proj\kidney_disease - kidney_disease.csv')
```

```python
# show all columns
pd.set_option('display.max_columns', None)
kidney_df
```

```python
print(kidney_df.head())
print(kidney_df.info())
print(kidney_df.describe())
print("Columns:", kidney_df.columns)
print("Shape:", kidney_df.shape)
```

```python
# checking for any missing values
print("Missing values in each column:")
print(kidney_df.isnull().sum())
```

```python
# Convert object to numeric types, coercing errors to NaN

for col in ['pcv', 'wc', 'rc']:
    kidney_df[col] = pd.to_numeric(kidney_df[col], errors='coerce')
```

```python
kidney_df.info()
```

```python
# Binary mapping for categorical columns
binary_map = {'yes': 1, 'no': 0, 'normal': 1, 'abnormal': 0,
              'present': 1, 'notpresent': 0, 'good': 1, 'poor': 0}
for col in ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane']:
    kidney_df[col] = kidney_df[col].map(binary_map)
```

```python
kidney_df
```

```python
# Target encoding
target_map = {'ckd': 1, 'notckd': 0}
kidney_df['classification'] = kidney_df['classification'].map(target_map)
# Display the updated DataFrame
kidney_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(5,4))
sns.countplot(x='classification', data=kidney_df, hue='classification',
              palette='Set2', legend=False)
plt.title("Class Distribution (CKD vs Not CKD)")
plt.xlabel("CKD Status (0 = Not CKD, 1 = CKD)")
plt.ylabel("Count")
plt.show()
```

```python
num_cols = ['age', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo']
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(data=kidney_df, x=col, hue="classification", kde=True, palette="Set2", bins=30)
    plt.title(f"Distribution of {col} by CKD Status")
    plt.show()
```

```python
# Fill missing values with median for numeric columns

for col in kidney_df.select_dtypes(include=['float64', 'int64']).columns:
    kidney_df[col] = kidney_df[col].fillna(kidney_df[col].median())
```

```python
# Display remaining missing values
print("Missing values after filling with median:")
print(kidney_df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
corr = kidney_df.corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()

# corelation details
corr_details = corr['classification'].sort_values(ascending=False)
print("Correlation with CKD classification:")
print(corr_details)
```

```python
# Drop id column
kidney_df = kidney_df.drop(columns=['id'])
```

```python
# Feature Target Split
x = kidney_df.drop('classification', axis=1)
y = kidney_df['classification']
```

```python
# Standard scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=42, max_iter=1000)
lr.fit(x_train, y_train)
# Predictions
y_pred = lr.predict(x_test)
```

```python
# logistic regression report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
# Confusion Matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not CKD', 'CKD'], yticklabels=['Not CKD', 'CKD'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# ROC Curve
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, lr.predict_proba(x_test)[:, 1])
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label='ROC curve (area ={:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
# print feature columns
print("Feature columns:", x.columns.tolist())
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'class_weight': ['balanced', None]
}

# Random Forest with GridSearch
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(
    estimator=rf, 
    param_grid=param_grid, 
    cv=5, 
    scoring='f1',   # f1 works well for imbalanced medical data
    n_jobs=-1, 
    verbose=2
)

grid_search.fit(x_train, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)

# Best model
best_rf = grid_search.best_estimator_

# Predictions
y_pred_best = best_rf.predict(x_test)

# Classification Report
from sklearn.metrics import classification_report, confusion_matrix
print("Tuned Random Forest Classification Report:")
print(classification_report(y_test, y_pred_best))

# Confusion Matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm_best = confusion_matrix(y_test, y_pred_best)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_best, annot=True, fmt='d', cmap='Oranges', xticklabels=['Not CKD', 'CKD'], yticklabels=['Not CKD', 'CKD'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Tuned Random Forest Confusion Matrix')
plt.show()
# ROC Curve for Tuned Random Forest
fpr_best, tpr_best, thresholds_best = roc_curve(y_test, best_rf.predict_proba(x_test)[:, 1])
roc_auc_best = auc(fpr_best, tpr_best)
plt.figure(figsize=(8, 6))
plt.plot(fpr_best, tpr_best, color='green', label='ROC curve (area ={:.2f})'.format(roc_auc_best))
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for Tuned Random Forest')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

# Cross-validation with 5 folds
cv_scores = cross_val_score(best_rf, x_scaled, y, cv=5, scoring='f1')

print("Cross-validation F1 scores:", cv_scores)
print("Mean F1 score:", cv_scores.mean())
```
